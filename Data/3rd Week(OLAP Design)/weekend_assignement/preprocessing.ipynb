```python
import psycopg2

def connect(database_name):
    return psycopg2.connect(user="postgres",
                                  password="admin",
                                  host="localhost",
                                  port="5432",
                                  database=database_name
                                  )
```

```python
def execute_select_query(query,connection):
    """This is the method to execute the select sql query given the parameter the query and connection method"""
    try:
        conn=connection
        #print(conn)
        cur=conn.cursor()
        cur.execute(query)
        data=cur.fetchall()

    except(Exception) as e:
        print(e)
    finally:
        if(conn):
            cur.close()
            conn.close()
            return data
```

```python
query="""
SELECT 
e.product_id,
p.product_name,
e.avg_price,
e.total_quantity as total_quantity_sold,
e.total_sales,
b.name as brand_name,
c.name as category
FROM 
each_product_total_sale e
INNER JOIN fact_product p
ON e.product_id = p.product_id
INNER JOIN dim_brand b
ON p.brand_id = b.id
INNER JOIN dim_category c
ON p.category_id = c.id
"""
```

```python
data=execute_select_query(query,connect('ecomm_warehouse'))
```
